```python
from pyspark.sql.functions import *
# from pyspark.sql.streaming import Trigger

# Path to your S3 bucket
raw_path = "s3://my-iot-data-bucket-2025/iot-data/"

bronze_path = "s3://my-iot-data-bucket-2025"

bronze_df = (spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "json")
    .option("cloudFiles.schemaLocation", bronze_path + "/schema/iot")
    .load(raw_path)
)

bronze_df.writeStream.format("delta") \
    .option("checkpointLocation", bronze_path + "/checkpoints/iot") \
    .outputMode("append") \
    .trigger(availableNow=True) \
    .table("bronze_iot_data")
```
